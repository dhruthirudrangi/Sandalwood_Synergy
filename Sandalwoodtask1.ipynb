```python
!pip install -U openai-whisper
!apt-get install -y ffmpeg
import whisper

# Load the Whisper model
model = whisper.load_model("small")  # You can specify "tiny", "base", "small", "medium", or "large"
print("Whisper model loaded successfully!")
!pip install librosa soundfile noisereduce
# Step 0: Install necessary packages in Google Colab
!pip install -U openai-whisper librosa soundfile noisereduce moviepy
!apt-get install -y ffmpeg  # Make sure ffmpeg is install

# Mount Google Drive to access files
from google.colab import drive
drive.mount('/content/drive')

import os
import whisper
import librosa
import soundfile as sf
import noisereduce as nr
from moviepy.editor import AudioFileClip

# Define the folders
input_folder = "/content/drive/My Drive/Sandalwood/your_mp3_folder"  # Path to MP3 files in Google Drive
wav_folder = "/content/drive/My Drive/Sandalwood/wav_folder"  # Folder to save WAV files
output_file = "/content/drive/My Drive/Sandalwood/translations.txt"  # File to save translations

# Create the WAV folder if it doesn't exist
os.makedirs(wav_folder, exist_ok=True)

# Step 1: Convert MP3 to WAV
def convert_mp3_to_wav(input_folder, wav_folder):
    for filename in os.listdir(input_folder):
        if filename.endswith(".mp3"):
            mp3_path = os.path.join(input_folder, filename)
            wav_path = os.path.join(wav_folder, filename.replace(".mp3", ".wav"))
            # Load MP3 and convert to WAV
            audio_clip = AudioFileClip(mp3_path)
            audio_clip.write_audiofile(wav_path, codec='pcm_s16le')
            audio_clip.close()
            print(f"Converted {filename} to WAV.")

# Step 2: Preprocess audio files and remove noise
def preprocess_audio_files(input_folder, output_folder):
    os.makedirs(output_folder, exist_ok=True)
    for filename in os.listdir(input_folder):
        if filename.endswith(".wav"):
            filepath = os.path.join(input_folder, filename)
            # Load audio file
            audio, sr = librosa.load(filepath, sr=16000)  # Load and resample to 16kHz
            # Apply noise reduction
            audio_denoised = nr.reduce_noise(y=audio, sr=sr)
            # Save the denoised audio
            output_path = os.path.join(output_folder, filename)
            sf.write(output_path, audio_denoised, 16000)  # Save in 16kHz mono format
            print(f"Processed {filename} with noise reduction.")

```

```python
# Step 3: Transcribe and Translate using Whisper
def transcribe_and_translate(input_folder, output_file):
    model = whisper.load_model("small")  # Load the Whisper model
    with open(output_file, "w", encoding="utf-8") as f:
        for filename in os.listdir(input_folder):
            if filename.endswith(".wav"):
                file_path = os.path.join(input_folder, filename)
                # Transcribe and translate the audio
                result = model.transcribe(file_path, task="translate")
                # Write the filename and the translated text to the output file
                f.write(f"{filename}: {result['text']}\n")
                print(f"Translated {filename} to English.")

# Run the conversion, preprocessing, and translation
convert_mp3_to_wav(input_folder, wav_folder)
preprocess_audio_files(wav_folder, wav_folder)  # Process WAV files in the same folder
transcribe_and_translate(wav_folder, output_file)
```
